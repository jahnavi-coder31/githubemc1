```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.read_csv("Walmart.csv")
df.head()
```

```python

df['Date'] = pd.to_datetime(df['Date'], format="%d-%m-%Y")
df.sort_values('Date', inplace=True)
```

```python

print("🔹 Dataset Info:")
print(df.info())
print("\n🔹 Summary Statistics:")
print(df.describe())
```

```python

plt.figure(figsize=(12, 6))
df.groupby('Date')['Weekly_Sales'].sum().plot()
plt.title('📈 Total Weekly Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Weekly Sales')
plt.grid(True)
plt.show()
```

```python

avg_sales = df.groupby('Store')['Weekly_Sales'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
avg_sales.plot(kind='bar', color='skyblue')
plt.title("🏪 Average Weekly Sales by Store")
plt.xlabel("Store")
plt.ylabel("Average Weekly Sales")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(8, 5))
sns.boxplot(x='Holiday_Flag', y='Weekly_Sales', data=df)
plt.title("🎉 Holiday vs Non-Holiday Weekly Sales")
plt.xlabel("Holiday (0 = No, 1 = Yes)")
plt.ylabel("Weekly Sales")
plt.show()
```

```python

df.describe().to_csv("summary_report.csv")
print("📁 Summary report saved as 'summary_report.csv'")
```

```python

top_sales_dates = df.groupby('Date')['Weekly_Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
top_sales_dates.plot(kind='bar')
plt.title("🔥 Top 10 Dates with Highest Total Sales")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

total_sales_by_store = df.groupby('Store')['Weekly_Sales'].sum().sort_values(ascending=False)
top_store = total_sales_by_store.idxmax()
print(f"🏆 Store {top_store} has the highest total sales: {total_sales_by_store.max():,.2f}")
```

```python

plt.figure(figsize=(10, 5))
sns.histplot(df['Weekly_Sales'], bins=50, kde=True)
plt.title("📊 Distribution of Weekly Sales")
plt.xlabel("Weekly Sales")
plt.ylabel("Frequency")
plt.show()
```

```python

df['Month'] = df['Date'].dt.month
monthly_sales = df.groupby('Month')['Weekly_Sales'].mean()

plt.figure(figsize=(10, 5))
monthly_sales.plot(kind='bar', color='orange')
plt.title("📅 Average Weekly Sales by Month")
plt.xlabel("Month")
plt.ylabel("Average Sales")
plt.xticks(rotation=0)
plt.show()
```

```python

plt.figure(figsize=(8, 6))
sns.scatterplot(x='CPI', y='Weekly_Sales', data=df, hue='Holiday_Flag')
plt.title("💰 CPI vs Weekly Sales (Holiday Highlighted)")
plt.xlabel("CPI")
plt.ylabel("Weekly Sales")
plt.show()
```

```python

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Temperature', y='Weekly_Sales', data=df)
plt.title("🌡️ Temperature vs Weekly Sales")
plt.xlabel("Temperature (°F)")
plt.ylabel("Weekly Sales")
plt.show()
```

```python

df['Year'] = df['Date'].dt.year
df['Week'] = df['Date'].dt.isocalendar().week
df['Day'] = df['Date'].dt.day_name()


df[['Date', 'Year', 'Week', 'Day']].head()
```

```python

avg_sales_by_day = df.groupby('Day')['Weekly_Sales'].mean().reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
)

plt.figure(figsize=(10, 5))
avg_sales_by_day.plot(kind='bar', color='lightgreen')
plt.title('📆 Average Sales by Day of the Week')
plt.xlabel('Day')
plt.ylabel('Average Weekly Sales')
plt.xticks(rotation=45)
plt.show()
```

```python

weekly_sales = df.groupby('Date')['Weekly_Sales'].sum().resample('W').sum()


plt.figure(figsize=(12, 5))
weekly_sales.plot()
plt.title("🔮 Resampled Weekly Total Sales")
plt.xlabel("Date")
plt.ylabel("Weekly Sales")
plt.grid(True)
plt.show()
```

```python

weekly_sales_rolling = weekly_sales.rolling(window=4).mean()

plt.figure(figsize=(12, 5))
plt.plot(weekly_sales, label='Original', alpha=0.5)
plt.plot(weekly_sales_rolling, label='4-Week Rolling Average', color='red')
plt.title("📉 Sales Trend with Rolling Average")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```python

from statsmodels.tsa.seasonal import seasonal_decompose

result = seasonal_decompose(weekly_sales, model='additive')
result.plot()
plt.suptitle("📊 Time Series Decomposition", fontsize=16)
plt.tight_layout()
plt.show()
```
